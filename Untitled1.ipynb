```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('celia_session_record.csv',index_col=0)
```

```python
#select out the markov sessions
df = df[df.Markov == 1]

#parse out the different conditions
df_90 = df[((df['Left Reward Prob'] == 0.9) | (df['Right Reward Prob'] == 0.9))]
df_80 = df[((df['Left Reward Prob'] == 0.8) | (df['Right Reward Prob'] == 0.8))]
df_70 = df[((df['Left Reward Prob'] == 0.7) | (df['Right Reward Prob'] == 0.7))]

#print out number of sessions and trials for each condition
print('# of 90-10 sessions: {} trials: {}'.format(df_90.shape[0],df_90['No. Trials'].sum()))
print('# of 80-20 sessions: {} trials: {}'.format(df_80.shape[0],df_80['No. Trials'].sum()))
print('# of 70-30 sessions: {} trials: {}'.format(df_70.shape[0],df_70['No. Trials'].sum()))

#I want to use df later, so...
df_all = df.copy()
```

```python
dfs = [df_90,df_80,df_70]
labels = ['90','80','70']

for df,label in zip(dfs,labels):
    plt.hist(df['p(high Port)'].values,alpha=0.5,label=label,bins=20)

plt.legend(loc='upper left')
plt.title('Distribution of correct port chosen')
```
